```python
import sqlalchemy as db
#create engine

import urllib
params = urllib.parse.quote_plus("DRIVER={SQL Server};SERVER=TWINBANJ\MSSQLSERVERDEV;DATABASE=adamdb")

engine = db.create_engine("mssql+pyodbc:///?odbc_connect=%s" % params)

conn = engine.connect()
trans = conn.begin()
#trans means transaction. Its used to begin the connection for each transaction. e.g. trans.committ()
#conn is the connection. Its used to connection the engine to server, something pandas can do automatically. e.g. conn.execute()
```

```python
#this one works. Pay attention conn.execute, NOT engine.execute. Must have '' at the start and end of each row of the query. Must commit at the end and close connection.

conn.execute('USE [adamdb]'
             'DROP TABLE IF EXISTS "ODS_GP_FULL";'
               'CREATE TABLE "ODS_GP_FULL" ('
               '[DSS_KEY] [int] IDENTITY(1,1) NOT NULL,'
'[CODE] [varchar](8) NOT NULL,'
'[NAME] [varchar](50) NOT NULL,'
            ' [NATIONAL_GROUPING] [varchar](5) NULL,'
'[HIGH_LEVEL_HEALTH_GEOGRAPHY] [varchar](5) NOT NULL,'
'[ADDRESS_LINE_1] [varchar](35) NULL,'
'[ADDRESS_LINE_2] [varchar](35) NULL,'
'[ADDRESS_LINE_3] [varchar](35) NULL,'
'[ADDRESS_LINE_4] [varchar](35) NULL,'
'[ADDRESS_LINE_5] [varchar](35) NULL,'
             '[POSTCODE] [varchar](8) NULL,'
'[OPEN_DATE] [date] NOT NULL,'
'[CLOSE_DATE] [date] NULL,'
'[STATUS_CODE] [varchar](1) NOT NULL,'
'[ORGANISATION_SUB_TYPE_CODE] [varchar](1) NOT NULL,'
'[PARENT_ORGANISATION_CODE] [varchar](6) NOT NULL,'
'[JOIN_PARENT_DATE] [date] NOT NULL,'
'[LEFT_PARENT_DATE] [date] NULL,'
'[CONTACT_TELEPHONE_NUMBER] [varchar](12) NULL,'
'[CURRENT_CARE_ORGANISATION] [varchar](5) NULL,'
'[DSS_RECORD_START_DATE] [date] NOT NULL,'
'[DSS_RECORD_END_DATE] [date] NULL,'
'[DSS_SYSTEM_CREATED_DATE] [date] NOT NULL,'
'[DSS_SYSTEM_UPDATED_DATE] [date] NOT NULL,'
                'CONSTRAINT [PK_ODS_GP_V02_DSS_KEY] PRIMARY KEY CLUSTERED'
'([DSS_KEY] ASC)WITH (PAD_INDEX = OFF, STATISTICS_NORECOMPUTE = OFF, IGNORE_DUP_KEY = OFF, ALLOW_ROW_LOCKS = ON, ALLOW_PAGE_LOCKS = ON) ON [PRIMARY],'
' CONSTRAINT [UK_ODS_GP_V02_CODE_DSS_RECORD_END_DATE] UNIQUE NONCLUSTERED'
'([CODE] ASC,[DSS_RECORD_END_DATE] ASC'
')WITH (PAD_INDEX = OFF, STATISTICS_NORECOMPUTE = OFF, IGNORE_DUP_KEY = OFF, ALLOW_ROW_LOCKS = ON, ALLOW_PAGE_LOCKS = ON) ON [PRIMARY],'
 'CONSTRAINT [UK_ODS_GP_V02_CODE_DSS_RECORD_START_DATE] UNIQUE NONCLUSTERED'
'([CODE] ASC,[DSS_RECORD_START_DATE] ASC'
')WITH (PAD_INDEX = OFF, STATISTICS_NORECOMPUTE = OFF, IGNORE_DUP_KEY = OFF, ALLOW_ROW_LOCKS = ON, ALLOW_PAGE_LOCKS = ON) ON [PRIMARY]'
') ON [PRIMARY];')

trans.commit()
conn.close()
```
